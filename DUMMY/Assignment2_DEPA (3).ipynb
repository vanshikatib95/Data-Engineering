```python
# a) List the actors (firstName, lastName) who acted in more then 25 movies.


sql= "SELECT first_name, last_name, count(film_id)\
FROM actor \
LEFT JOIN film_actor \
using (actor_id) \
GROUP BY  actor_id \
HAVING  count(film_id) > 25;"
qry(sql)
```

```python
 #b) List the actors who have worked in the German language movies.
    
sql = "SELECT DISTINCT a.first_name, a.last_name \
FROM actor AS a  \
INNER JOIN film_actor AS b \
ON (a.actor_id)=(b.actor_id) \
INNER JOIN film AS c \
on (b.film_id)=(c.film_id) \
WHERE  c.film_id = 6;"

qry(sql)
```

```python
# c) List the actors who acted in horror movies.
# Note: Show the count of movies against each actor in the result set.

sql = "                                         \
SELECT                                          \
	a.actor_id,                                 \
    a.first_name,                               \
    a.last_name,                                \
    COUNT(*) AS horror_movies                   \
FROM                                            \
	actor a                                     \
JOIN film_actor f ON a.actor_id=f.actor_id      \
JOIN film_category fc ON f.film_id=fc.film_id   \
JOIN category c ON c.category_id=fc.category_id \
WHERE c.name='Horror'                           \
GROUP BY a.actor_id;"

qry(sql)
```

```python
# d) List all customers who rented more than 3 horror movies.

sql= "                                                      \
SELECT                                                      \
DISTINCT(h.customer_id),                                    \
h.first_name,                                               \
h.last_name,                                                \
COUNT(*) AS horror_Movies                                   \
FROM                                                        \
customer h                                                  \
JOIN rental r ON h.customer_id=r.customer_id                \
LEFT JOIN inventory i ON r.inventory_id=i.inventory_id      \
JOIN film_category fg ON i.film_id=fg.film_id               \
JOIN category c ON c.category_id=fg.category_id             \
WHERE c.name='Horror'                                       \
GROUP BY h.customer_id                                      \
HAVING horror_Movies > 3;"
qry(sql)
```

```python
# e) List all customers who rented the movie which starred SCARLETT BENING
sql= "SELECT DISTINCT \
	a.customer_id, \
    a.first_name \
FROM \
	customer a \
JOIN rental r ON a.customer_id=r.customer_id \
JOIN inventory i ON r.inventory_id=i.inventory_id \
JOIN film f ON i.film_id=f.film_id \
JOIN film_actor ff ON f.film_id = ff.film_id \
JOIN actor aa ON aa.actor_id = ff.actor_id \
WHERE aa.first_name='SCARLETT' AND aa.last_name='BENING';"
qry(sql)
```

```python
# f) Which customers residing at postal code 62703 rented movies that were Documentaries. 
sql= "SELECT \
c.customer_id, \
c.first_name, \
c.last_name \
FROM \
customer c \
JOIN address a ON c.address_id=a.address_id \
JOIN rental r ON c.customer_id=r.customer_id \
JOIN inventory i ON r.inventory_id=i.inventory_id \
JOIN film_category fg ON i.film_id=fg.film_id \
JOIN category g ON g.category_id=fg.category_id \
WHERE g.name='Documentary' AND a.postal_code = '62703' \
GROUP BY c.customer_id;"

qry(sql)
```

```python
# j) Find all the film categories in which there are between 55 and 65 films. 
#Return the names of categories and the
#number of films per category, sorted from highest to lowest by the number of films. 

sql= "SELECT \
c.name, \
COUNT(f.film_id) AS FilmNumber \
FROM \
film_category f \
JOIN \
category c ON f.category_id=c.category_id \
GROUP BY f.category_id \
HAVING FilmNumber  BETWEEN 55 AND 65 \
ORDER BY FilmNumber DESC;"

qry(sql)
```

```python
# k) In which of the film categories is the average difference 
#between the film replacement cost and the rental rate larger than 17$?

sql = "SELECT \
fc.category_id, \
c.name, \
AVG(f.replacement_cost-f.rental_rate) AS AVGdiff \
FROM \
film_category fc \
JOIN film f ON fc.film_id=f.film_id \
JOIN category c ON fc.category_id = c.category_id \
GROUP BY fc.category_id \
HAVING AVGdiff > 17;"

qry(sql)
```

```python
# l) Many DVD stores produce a daily list of overdue rentals so that customers can be contacted 
#and asked to return their overdue DVDs. To create such a list, 
#search the rental table for films with a return date that is NULL and 
#where the rental date is further in the past than the rental duration specified in the film table. 
#If so, the film is overdue and we should produce the name of the film along with the customer name 
#and phone number.

sql = "SELECT         \
    c.first_name,  \
    c.last_name,   \
    a.phone,   \
    f.title,   \
    r.rental_date,   \
    f.rental_duration,  \
    r.return_date, \
    DATE_ADD(r.rental_date,    \
        INTERVAL f.rental_duration DAY) AS est_return_date   \
FROM     \
    rental r     \
        INNER JOIN     \
    customer c ON r.customer_id = c.customer_id    \
        INNER JOIN     \
    address a ON c.address_id = a.address_id   \
        INNER JOIN      \
    inventory i ON r.inventory_id = i.inventory_id     \
        INNER JOIN        \
    film f ON i.film_id = f.film_id      \
WHERE       \
    r.return_date IS NULL     \
        OR r.return_date > DATE_ADD(r.rental_date,     \
        INTERVAL f.rental_duration DAY);"

qry(sql)
```

```python
##5 a) Return top 10 customers by revenue

sql = "SELECT c.first_name, c.last_name, SUM(amount) AS Revenue \
FROM payment as p \
INNER JOIN sakila.customer as c ON p.customer_id=c.customer_id \
GROUP BY c.first_name, c.last_name \
ORDER BY Revenue DESC \
limit 10;"
qry(sql)
```

```python
# 5b) Return top categories in inventory by movie count
sql = "SELECT count(f.film_id), c.name as category \
FROM film f JOIN film_category fg ON f.film_id=fg.film_id \
JOIN category c ON c.category_id=fg.category_id \
Group by c.name \
order by count(f.film_id) desc;"
qry(sql)
```

```python
#5c) Identify Film categories that customers rent for most days -business can focus on promoting less-popular 
# categories (potentially by offering discounts)

sql = "SELECT c.name, SUM(TIMESTAMPDIFF(day,rental_date,return_date)) AS total_rental_days \
FROM sakila.category as c \
INNER JOIN sakila.film_category as fc ON fc.category_id=c.category_id \
INNER JOIN sakila.inventory as i ON fc.film_id=i.film_id \
INNER JOIN sakila.rental as r ON i.inventory_id=r.inventory_id \
GROUP BY c.name \
ORDER BY total_rental_days DESC"

qry(sql)
```

```python
#5d) # Identify laguages which make the most revenue - business can promote movies of less-popular films

sql = "SELECT l.name, SUM(p.amount) as revenue \
FROM sakila.language as l \
INNER JOIN sakila.film as f ON l.language_id=f.language_id \
INNER JOIN sakila.inventory as i ON f.film_id=i.film_id \
INNER JOIN sakila.rental as r ON i.inventory_id=r.inventory_id \
INNER JOIN sakila.payment as p ON r.rental_id=p.rental_id \
GROUP BY l.name \
ORDER BY revenue DESC"

qry(sql)
```
